```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as st
plt.rcParams.update({"font.size": 14, "axes.titlesize": 16, "axes.labelsize": 16, "axes.labelweight": "bold", "xtick.labelsize": 14, "ytick.labelsize": 14, "legend.fontsize": 14, "figure.titlesize": 16, "figure.figsize": (8, 6)})
```

```python
x = np.linspace(-10, 10, 1000)
y1 = st.norm.pdf(x, 0, 1)
ypt5 = st.norm.pdf(x, 0, .5)

y2 = st.norm.pdf(x, 0, 2)
y3 = st.norm.pdf(x, 0, 3)
```

```python
plt.xkcd()
plt.figure(figsize=(16,9))
plt.plot(x, y1, label="N(0,1)")
plt.plot(x, y2, label="N(0,2)")
plt.plot(x, ypt5, label="N(0,0.5)")
plt.plot(x, y3, label="N(0,3)")
plt.axvline(0, color = "k", label = "Mean = 0", ls ="--")
plt.legend()
plt.xlabel("x")
plt.ylabel("Probability Density")
plt.title("Normal Distribution with Different Standard Deviations")
plt.grid()
plt.savefig("normal_dist_std.png", dpi=600)
```

```python
plt.xkcd()
x1 = np.linspace(-1, 1, 1000)
x2 = np.linspace(-2, 2, 1000)
yb1 = st.norm.pdf(x1, 0, 1)
yb2 = st.norm.pdf(x2, 0, 2)
plt.figure(figsize=(16,9))
plt.plot(x, y1, label="N(0,1)")
plt.fill_between(x1, yb1, alpha=0.5)
plt.plot(x, y2, label="N(0,2)")
plt.fill_between(x2, yb2, alpha=0.5)
plt.axvline(0, color = "k", label = "Mean = 0", ls ="--")
plt.legend()
plt.xlabel("x")
plt.ylabel("Probability Density")
plt.ylim(bottom = 0)
plt.title("Normal Distribution with Different Standard Deviations")
plt.savefig("normal_dist_std_fill.png", dpi=300)
# plt.grid()
```

```python
dt = 0.05
T = 10
N = int(T/dt)
x = np.zeros(N)
x[0] = 0
D = 1
def brownian_motion(x, D, dt):
    for i in range(1, N):
        x[i] = x[i-1] + 2*D*np.random.normal(0, dt) * np.sqrt(dt)
    return x
x = brownian_motion(x, D, dt)
# for i in range(1, N):
#     x[i] = x[i-1] + 2*D*np.random.normal(0, dt) * np.sqrt(dt)

t = np.linspace(0, T, N)

plt.figure(figsize=(16,9))
plt.scatter(t,x)
plt.plot(t, x, label="Brownian Motion", zorder=0)
# plt.axhline(0, color = "k", label = "Mean = 0", ls ="--")
plt.xlabel("t")
plt.ylabel("X(t)")
plt.title("Brownian Motion with Different Standard Deviations")
plt.legend()
# plt.grid()
plt.savefig("brownian_motion.png", dpi=300)
plt.show()
```
