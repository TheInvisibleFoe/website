```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
```

```python
fig = plt.figure()
axes = plt.axes(projection = '3d')
R = 5
def x(u,v):
    return (R+u*np.cos(v/2))*np.cos(v)
def y(u,v):
    return (R+u*np.cos(v/2))*np.sin(v)
def z(u,v):
    return u*np.sin(v/2)
t = np.linspace(0,2*np.pi,500)
s = np.linspace(-R,R,5)
axes.set_zlim(-10,10)

for i in s:
    plt.plot(x(i,t),y(i,t),z(i,t))



```
